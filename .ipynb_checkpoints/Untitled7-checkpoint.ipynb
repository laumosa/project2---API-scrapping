```python
import requests
import pandas as pd

# Set API endpoint URL
url = 'https://api.openaq.org/v1/measurements'

# Set query parameters
params = {
    'country': 'ES',
    'city': 'Barcelona',
    'parameter': 'pm25',
    'limit': 10000,
    'page': 1
}

# Initialize empty list to store all results
all_results = []

# Loop through pages of API response
while True:
    # Send API request for current page
    response = requests.get(url, params=params)
    response_json = response.json()
    
    # Add results for current page to all_results list
    all_results.extend(response_json['results'])
    
    # Check if there are more pages of results
    if 'next' in response_json['meta'] and response_json['meta']['next'] is not None:
        # If there are more pages, update 'page' parameter and continue loop
        params['page'] += 1
    else:
        # If there are no more pages, break out of loop
        break

# Convert results to pandas DataFrame
df = pd.json_normalize(all_results)

# Print DataFrame
print(df)
```
