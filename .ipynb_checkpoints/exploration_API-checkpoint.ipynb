## 1. Import libraries

```python
import requests 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
```

## 2. API request

### First API request:

```python
# Set API endpoint URL
url = 'https://api.openaq.org/v1/measurements'

# Set query parameters
params = {
    'country': 'ES',
    'city': 'Madrid',
    'parameter': 'pm25',
    'limit': 10000,    
    'date_from': '2020-01-01T00:00:00Z',
    'date_to': '2020-05-31T23:59:59Z',
    'page': 1,
}

# Send initial API request
response = requests.get(url, params=params)
response_json_1 = response.json()
```

```python
df_1 = pd.DataFrame(response_json_1['results'])
df_1
```

### Second API request:

```python
# Set API endpoint URL
url = 'https://api.openaq.org/v1/measurements'

# Set query parameters
params = {
    'country': 'ES',
    'city': 'Madrid',
    'parameter': 'pm25',
    'limit': 10000,    
    'date_from': '2020-01-01T00:00:00Z',
    'date_to': '2020-05-31T23:59:59Z',
    'page': 2,
}

# Send initial API request
response = requests.get(url, params=params)
response_json_2 = response.json()
```

```python
df_2 = pd.DataFrame(response_json_2['results'])
df_2
```

## 3. Join Dataframes

```python
df_concat = pd.concat([df_1, df_2], axis=0, ignore_index=True)
```

## 5. General cleaning:

```python
# Extract information from dictionaries
df_concat['date_utc'] = df_concat['date'].apply(lambda x: x['utc'])
df_concat['date_local'] = df_concat['date'].apply(lambda x: x['local'])
df_concat['coordinates_latitude'] = df_concat['coordinates'].apply(lambda x: x['latitude'])
df_concat['coordinates_longitude'] = df_concat['coordinates'].apply(lambda x: x['longitude'])
```

```python
# Drop dictionaries
df_concat.drop(columns=["date", "coordinates"], axis = 1, inplace = True)
```

```python
# Create new colum for year, month and day, respectively
df_concat['year'] = df_concat['date_utc'].apply(lambda x: int(x[:4]))
df_concat['month'] = df_concat['date_utc'].apply(lambda x: int(x[6]))
df_concat['day'] = df_concat['date_utc'].apply(lambda x: int(x[8:10]))
```

```python
pd.isna(df_concat).sum()
```

```python
df_concat.duplicated().sum()
```

```python
# Replace each value in the 'Month' column with a string representing the corresponding month
months_dict = {1: 'January', 2: 'February', 3: 'March (Covid)', 4: 'April', 5: 'May'}
df_concat['month_name'] = df_concat['month'].replace(months_dict)
```

```python
df_concat
```

## 6. Plot:

```python
# Set plot style
sns.set_style('whitegrid')

# Create the plot with hue for each year
ax = sns.barplot(data=df_concat, x='month_name', y='value', linewidth=2.5, ci=None, color=sns.color_palette('coolwarm')[1])

# Add plot title and axis labels
ax.set_title('Average levels of pm25 in Madrid (Jan-May 2020)', fontsize=16)
ax.set_xlabel('Month', fontsize=14)
ax.set_ylabel('Average levels of pm25', fontsize=14)

# Set tick label font size and rotation
ax.tick_params(labelsize=10, rotation=45)

# Invert x-tick labels
ax.set_xticklabels(ax.get_xticklabels()[::-1])

# Set the color for the March label
ax.get_xticklabels()[2].set_color('red')

# Display the plot
plt.show()
```

```python
# Create the plot with hue for each year
ax = sns.boxplot(data=df_concat, x='month_name', y='value', hue='month_name', color=sns.color_palette('coolwarm')[1], dodge=False, width=0.5)

# Add plot title and axis labels
ax.set_title('Average levels of pm25 in Madrid (Jan-May 2020)', fontsize=16)
ax.set_xlabel('Month', fontsize=14)
ax.set_ylabel('Average levels of pm25', fontsize=14)

# Remove legend
plt.gca().get_legend().remove()

# Invert x-tick labels
ax.set_xticklabels(ax.get_xticklabels()[::-1])

# Display the plot
plt.show()
```
