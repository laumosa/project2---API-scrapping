```python
import requests
import pandas as pd

# Set API endpoint URL
url = 'https://api.openaq.org/v1/measurements'

# Set query parameters
params = {
    'country': 'ES',
    'city': 'Madrid',
    'parameter': 'pm25',
    'limit': 100,    
    'date_from': '2020-02-01T00:00:00Z',
    'date_to': '2020-04-30T23:59:59Z',
}

# Send initial API request
response = requests.get(url, params=params)
response_json = response.json()

# Get total number of pages
total_pages = response_json['meta']['found'] // params['limit'] + 1

# Initialize an empty list to store results
results_list = []

# Loop through all pages of results
for page in range(1, total_pages + 1):
    # Set page parameter
    params['page'] = page
    
    # Send API request for current page
    response = requests.get(url, params=params)
    response_json = response.json()
    
    # Process results for current page
    results = response_json['results']
    for result in results:
        # Append result to results_list
        results_list.append(result)

# Convert results_list to a Pandas DataFrame
df = pd.DataFrame(results_list)

# Print the DataFrame
df
```
